```python
# Let's import our libraries.
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Auto dataset
df = pd.read_csv("Boston.csv")
df
# Now let's apply K-Means Clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(df.drop(['medv'], axis=1))

# Now let's add the clusters label to our dataset
df['cluster'] = kmeans.labels_

# We can now visualize the clustered data
plt.figure(figsize=(6, 6))
plt.scatter(df['rm'], df['ptratio'], c=df['cluster'], cmap='viridis')
plt.title('Boston Dataset Clustered using K-Means')
plt.xlabel('Average Number of Rooms')
plt.ylabel('Pupil-Teacher Ratio by Town')
plt.show()
```

```python
# Let's import our libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram


# We can now load our dataset
df = pd.read_csv("Boston.csv")
df

# Let's compute our linkage matrix
Z = linkage(df.drop(['medv'], axis=1),method="ward")

# We can now plot the dendrogram
fig = plt.figure(figsize=(8, 5))
dn = dendrogram(Z)
plt.title('Boston Dataset Hierarchical Clustering Dendrogram')
plt.show()
```

```python
# Let's import our Libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

# we can now load our dataset
df = pd.read_csv("Boston.csv")

# we can now seperate the features and target 
X = df.drop(['medv'], axis=1)
y = df['medv']

# Let's initialize the PCA model
pca = PCA()

# let's fit the model in our dataset
pca.fit(X)

# we can now print our explained variable ratio
print("Explained Variance Ratio: ", pca.explained_variance_ratio_)

# let's now transform the dataset using PCA
X_pca = pca.transform(X)

# Let's print the shape of the transform data
print("shape of transformed dataset:", X_pca.shape)
```
