```python
# We gonna Import our Libraries
import numpy as np
from matplotlib.pyplot import subplots, cm
import sklearn.model_selection as skm
from ISLP import load_data, confusion_table
from sklearn.svm import SVC
from ISLP.svm import plot as plot_svm
from sklearn.metrics import RocCurveDisplay
```

```python
# Let's use the function RocCurveDisplay.from_estimator to produce several ROC plots, using a shorthand roc_curve.
roc_curve = RocCurveDisplay.from_estimator # shorthand
```

```python
rng = np.random.default_rng(1)
X = rng.standard_normal((50, 2))
y = np.array([-1]*25+[1]*25)
X[y==1] += 1
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0],
           X[:,1],
           c=y,
           cmap=cm.coolwarm);
```

```python
# We now fit the classifier
svm_linear = SVC(C=10, kernel="linear")
svm_linear.fit(X, y)
```

```python
# We now plot the vector classifier with two features by visualising values of its Decision function.
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
        y,
        svm_linear,
        ax=ax)
```

```python
# Let's try to fit in a much smaller cost parameter.
svm_linear_small = SVC(C=0.1, kernel="linear")
svm_linear_small.fit(X, y)
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
         y,
         svm_linear_small,
         ax=ax)
```

```python
# Now let's try to extract the coefficients of the linear decision boundary.
svm_linear.coef_
```

```python
# Now let's use the usual machinery to tune it, since our svm is an estimator in sklearn
kfold = skm.KFold(5,
                 random_state=0,
                 shuffle=True)
grid = skm.GridSearchCV(svm_linear,
                        {'C':[0.001,0.01,0.1,1,5,10,100]},
                        refit=True,
                        cv=kfold,
                        scoring='accuracy')
grid.fit(X, y)
grid.best_params_
```

```python
# Now let access the cross-validation errors for each of these models.
grid.cv_results_[("mean_test_score")]
```

```python
# Let's generate a test data set
X_test = rng.standard_normal((20, 2))
y_test = np.array([-1]*10+[1]*10)
X_test[y_test==1] += 1
```

```python
# Now we predict the class labels of these observations.
best_ = grid.best_estimator_
y_test_hat = best_.predict(X_test)
confusion_table(y_test_hat, y_test)
```

```python
# Let's use a value of 0.001 for C
svm_ = SVC(C=0.001,
           kernel='linear').fit(X, y)
y_test_hat = svm_.predict(X_test)
confusion_table(y_test_hat, y_test)
```

```python
# Let's consider a situation in which the two classes are linearly seperable.
X[y==1] += 1.9;
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0], X[:,1], c=y, cmap=cm.coolwarm);
```

```python
# The observations are linearly separable.
svm_ = SVC(C=1e5, kernel="linear").fit(X, y)
y_hat = svm_.predict(X)
confusion_table(y_hat, y)
```

```python
# We now fit the support vector classifier and plot the resulting hyperlane, using a large value of C.
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
        y,
        svm_,
        ax=ax)
```

```python
# Let's try a smaller value of C.
svm_ = SVC(C=0.1, kernel='linear').fit(X, y)
y_hat = svm_.predict(X)
confusion_table(y_hat, y)
```

```python
# it seems possible that this model will perform better on test data than the model with c=1e5
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
        y,
        svm_,
        ax=ax)
```

```python
# Now let's try to fit in an SVM using a non-linear kernel, we first generate some data with non-linear class boundary.
X = rng.standard_normal((200, 2))
X[:100] += 2
X[100:150] -= 2
y = np.array([1]*150+[4]*50)
```

```python
# Now let's plot our data.
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0],
           X[:,1],
           c=y,
           cmap=cm.coolwarm)
```

```python
# The data is randomly split into training and testing groups, we now fit the training data using the SVC() estimator wth a radial kernel and Gamma = 1.
(X_train, 
 X_test,
 y_train,
 y_test) = skm.train_test_split(X,
                                y,
                                test_size=0.5,
                                random_state=0)
svm_rbf = SVC(kernel="rbf", gamma=1, C=1)
svm_rbf.fit(X_train, y_train)
```

```python
 fig, ax = subplots(figsize=(8,8))
plot_svm(X_train,
         y_train,
         svm_rbf,
         ax=ax)
```

```python
# Since there is a fair number of training errors in this SVM, Let's increase the value of C to reduce the number of training errors.
svm_rbf = SVC(kernel="rbf", gamma=1, C=1e5)
svm_rbf.fit(X_train, y_train)
fig, ax = subplots(figsize=(8,8))
plot_svm(X_train,
         y_train,
         svm_rbf,
         ax=ax)
```

```python
# We can now perform cross-validation to select the best choice of Gamma and C for an SVM with a radial kernel.
kfold = skm.KFold(5, 
                  random_state=0,
                  shuffle=True)
grid = skm.GridSearchCV(svm_rbf,
                        {'C':[0.1,1,10,100,1000],
                         'gamma':[0.5,1,2,3,4]},
                        refit=True,
                        cv=kfold,
                        scoring='accuracy');
grid.fit(X_train, y_train)
grid.best_params_
```

```python
# The best choice of the parameters under five-fold CV are archived at C=1 and gamma=0.5.
best_svm = grid.best_estimator_
fig, ax = subplots(figsize=(8,8))
plot_svm(X_train,
         y_train,
         best_svm,
         ax=ax)

y_hat_test = best_svm.predict(X_test)
confusion_table(y_hat_test, y_test)
```

```python
# Now let's use the roc_curve() to produce a plot of ROC curve.
fig, ax = subplots(figsize=(8,8))
roc_curve(best_svm,
         X_train,
         y_train,
         name="Training",
         color="g",
         ax=ax);
```

```python
# The SVM produce accurate predictions, Now let's increase gamma to produce a more flexible fit.
svm_flex = SVC(kernel="rbf", 
              gamma=50,
              C=1)
svm_flex.fit(X_train, y_train)
fig, ax = subplots(figsize=(8,8))
roc_curve(svm_flex,
          X_train,
          y_train,
          name='Training $\gamma=50$',
          color='g',
          ax=ax);
```

```python
# Since the ROC curve are all on the training data, we really are more interestedin the level of prediction accuracy on the test data.Let's compute on the test data.
roc_curve(svm_flex,
          X_test,
          y_test,
          name='Test $\gamma=50$',
          color='b',
          ax=ax)
fig;
```

```python
# Let's look at our tuned SVM.
fig, ax = subplots(figsize=(8,8))
for (X_, y_, c, name) in zip(
     (X_train, X_test),
     (y_train, y_test),
     ('g', 'b'),
     ('CV tuned on training',
      'CV tuned on test')):
    roc_curve(best_svm,
              X_,
              y_,
              name=name,
              ax=ax,
              color=c)
```

```python
# Now let's use the SVM with multiple classes to explore that setting briefly here by generating a third class of observations.
rng = np.random.default_rng(123)
X = np.vstack([X, rng.standard_normal((50, 2))])
y = np.hstack([y, [0]*50])
X[y==0,1] += 2
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0], X[:,1], c=y, cmap=cm.coolwarm);
```

```python
# We now fit an SVM to the data.
svm_rbf_3 = SVC(kernel="rbf",
                C=10,
                gamma=1,
                decision_function_shape='ovo');
svm_rbf_3.fit(X, y)
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
         y,
         svm_rbf_3,
         scatter_cmap=cm.tab10,
         ax=ax)
```
